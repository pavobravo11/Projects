```python
import yfinance as yf
```

```python
bmo = yf.Ticker("BMO")
cenovus = yf.Ticker("CVE")
```

```python
bmo_last_year = bmo.history("1y")['Close']
cen_last_year = cenovus.history("1y")['Close']
```

```python
import matplotlib.pyplot as plt

fig, (ax1,ax2) = plt.subplots(2, 1, figsize=(12, 8))

ax1.plot(bmo_last_year)
ax1.set_title("BMO Daily Returns")
ax2.plot(cen_last_year, color="green")
ax2.set_title("Cenovus Daily Returns")

plt.show()
```
